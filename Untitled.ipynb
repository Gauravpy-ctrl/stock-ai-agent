```python
import yfinance as yf
import pandas as pd
import requests
```

```python
# STOCK DATA
def get_stock_data(symbol):
    stock = yf.Ticker(symbol)
    return stock.history(period="6mo")

# RSI
def calculate_rsi(data, window=14):
    delta = data['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window).mean()
    rs = gain / loss
    return 100 - (100 / (1 + rs))
```

```python
API_KEY = "d7q97b1r01qosaar34q0d7q97b1r01qosaar34qg"

from datetime import date, timedelta
import requests

def get_stock_news(symbol):
    today = date.today()
    past = today - timedelta(days=7)

    url = f"https://finnhub.io/api/v1/company-news?symbol={symbol}&from={past}&to={today}&token={API_KEY}"
    
    response = requests.get(url)
    data = response.json()

    # ✅ FIX: handle both list & error cases
    if isinstance(data, list):
        return data[:5]
    else:
        print("⚠️ API Error:", data)
        return []

def analyze_news_sentiment(news):
    positive = ["growth", "profit", "gain", "up"]
    negative = ["loss", "fall", "drop", "down"]

    score = 0
    for article in news:
        headline = article['headline'].lower()

        if any(word in headline for word in positive):
            score += 1
        if any(word in headline for word in negative):
            score -= 1

    if score > 0:
        return "Positive"
    elif score < 0:
        return "Negative"
    return "Neutral"
```

```python
def technical_agent(symbol):
    data = get_stock_data(symbol)

    data['RSI'] = calculate_rsi(data)
    data['MA50'] = data['Close'].rolling(50).mean()

    return {
        "price": data['Close'].iloc[-1],
        "rsi": data['RSI'].iloc[-1],
        "ma": data['MA50'].iloc[-1]
    }


def news_agent(symbol):
    news = get_stock_news(symbol.replace(".NS", ""))
    sentiment = analyze_news_sentiment(news)

    return {"sentiment": sentiment}


def decision_agent(tech, news):
    price, rsi, ma = tech["price"], tech["rsi"], tech["ma"]
    sentiment = news["sentiment"]

    if price > ma and rsi < 70 and sentiment == "Positive":
        return "BUY", "Strong trend + positive news"

    elif rsi > 70:
        return "HOLD", "Overbought"

    elif sentiment == "Negative":
        return "AVOID", "Negative sentiment"

    elif price < ma:
        return "AVOID", "Downtrend"

    return "HOLD", "Mixed signals"
```

```python
def stock_multi_agent(symbol):
    print(f"\nAnalyzing {symbol}...\n")

    tech = technical_agent(symbol)
    news = news_agent(symbol)
    decision, reason = decision_agent(tech, news)

    print("Price:", tech["price"])
    print("RSI:", tech["rsi"])
    print("Sentiment:", news["sentiment"])
    print("Decision:", decision)
    print("Reason:", reason)
```

```python
# Simple in-memory storage
memory = {
    "risk_level": "medium",
    "preferred_stocks": []
}
```

```python
def update_memory(risk=None, stock=None):
    if risk:
        memory["risk_level"] = risk
    
    if stock and stock not in memory["preferred_stocks"]:
        memory["preferred_stocks"].append(stock)
```

```python
def decision_agent(tech, news):
    price, rsi, ma = tech["price"], tech["rsi"], tech["ma"]
    sentiment = news["sentiment"]
    risk = memory["risk_level"]

    # 🔥 Personalized logic
    if risk == "low":
        if price > ma and rsi < 60 and sentiment == "Positive":
            return "BUY", "Low-risk safe entry"
        else:
            return "HOLD", "Not safe for low-risk profile"

    elif risk == "high":
        if rsi < 75:
            return "BUY", "Aggressive entry allowed"
    
    # Default logic
    if price > ma and rsi < 70 and sentiment == "Positive":
        return "BUY", "Strong trend + positive news"

    elif rsi > 70:
        return "HOLD", "Overbought"

    elif sentiment == "Negative":
        return "AVOID", "Negative sentiment"

    elif price < ma:
        return "AVOID", "Downtrend"

    return "HOLD", "Mixed signals"
```

```python
def stock_multi_agent(symbol):
    print(f"\nAnalyzing {symbol}...\n")

    # Update memory
    update_memory(stock=symbol)

    tech = technical_agent(symbol)
    news = news_agent(symbol)
    decision, reason = decision_agent(tech, news)

    print("Price:", tech["price"])
    print("RSI:", tech["rsi"])
    print("Sentiment:", news["sentiment"])
    print("Risk Level:", memory["risk_level"])
    print("Decision:", decision)
    print("Reason:", reason)
```

```python
import json
import os

MEMORY_FILE = "memory.json"

def load_memory():
    if os.path.exists(MEMORY_FILE):
        with open(MEMORY_FILE, "r") as f:
            return json.load(f)
    else:
        return {
            "risk_level": "medium",
            "preferred_stocks": []
        }

def save_memory(memory):
    with open(MEMORY_FILE, "w") as f:
        json.dump(memory, f, indent=4)

# Load once
memory = load_memory()
```

```python
def update_memory(risk=None, stock=None):
    if risk:
        memory["risk_level"] = risk

    if stock and stock not in memory["preferred_stocks"]:
        memory["preferred_stocks"].append(stock)

    save_memory(memory)  # 🔥 auto-save
```
